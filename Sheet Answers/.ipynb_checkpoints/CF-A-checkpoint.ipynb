# [Beautiful Matrix](https://codeforces.com/contest/263/my)

```python
matrix = []
t_matrix = []
current_row = 0
current_column = 0

for row in range(5):
    matrix.append([])
    matrix[row].extend(list(map(int, input().split())))
    if sum(matrix[row]) == 1:
        current_row = row + 1

t_matrix = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))] 

for column in range(5):
    if sum(t_matrix[column]) == 1:
        current_column = column + 1

output = abs(current_row - 3) + abs(current_column - 3)

print(output)
```

# [Gravity Flip](https://codeforces.com/contest/405/my)

```python
n = int(input())

columns = []
columns.extend(list(map(int, input().split())))
columns.sort()

for i in range(n):
    print(columns[i], end=" ")
```

```python
def swap(current, after):
    if columns[after] < columns[current]:
        columns[current], columns[after] = columns[after], columns[current]
    else:
        pass

    if current != 0:
        swap(current-1, after-1)
    
n = int(input())

columns = []
columns.extend(list(map(int, input().split())))

for i in range(n-1):
    swap(i, i+1)

for i in range(n):
    print(columns[i], end=" ")
```

# [Sereja & Dima](https://codeforces.com/contest/381/my)

```python
n = int(input())

def maximum(first, last):
    return max(cards[first], cards[last-1])

cards = []
cards.extend(list(map(int, input().split())))
sereja = 0
dima = 0
maximum_num = 0

for i in range(n):
    maximum_num = maximum(0, len(cards))
    maximum_index = cards.index(maximum_num)
    
    if i%2==0:
        dima = dima + maximum_num
    else:
        sereja = sereja + maximum_num

    cards.pop(maximum_index)
    
print(dima, sereja)
```

# [Police Recruits](https://codeforces.com/contest/427/my)

```python
n = int(input())

crime_list = list(map(int, input().split()))
officers = 0
untreated_crimes = 0

while crime_list[-1] > 0 and len(crime_list) > 1:
    crime_list.pop(-1)

for i in crime_list:
    if i > 0:
        officers += i
    else:
        if officers == 0:
            untreated_crimes = untreated_crimes + i
        else:
            officers = officers + i
    
print(abs(untreated_crimes))
```

# [Night at The Museum](https://codeforces.com/contest/731/my)

```python
string = input()
moves = 0
pointer = 0
index = 0
walk = 0

for i in range(len(string)):
    index = ord(string[i]) - 97
    walk = abs(pointer - index)
    if walk <= 13:
        moves += walk
    else:
        moves += 26 - walk
    
    pointer = index
    
print(moves)

```

#  [Die Roll](https://codeforces.com/contest/9/my)

```python
from fractions import Fraction

y, w = map(int, input().split())

bigger = y if y > w else w
numerator = 6 - (bigger - 1)
ratio = str(Fraction(numerator, 6))
if ratio == "1":
    print("1/1")
elif ratio == "0":
    print("0/1")
else:
    print("{}/{}".format(ratio[0], ratio[2]))
```

# [Shaass and Oskols](https://codeforces.com/contest/294/my)

```python
n = int(input())
boxes = list(map(int, input().split()))
m = int(input())
positions = []
for row in range(m):
    positions.append([])
    positions[row] = list(map(int, input().split()))

for i in range(m):
    
    if n == 1:
            boxes[positions[i][0]-1] = 0
    elif n == 2:
        if positions[i][0] == 1:
            boxes[positions[i][0]] += boxes[positions[i][0]-1] - positions[i][1]
            boxes[positions[i][0]-1] = 0
        elif positions[i][0] == n:
            boxes[positions[i][0]-1-1] += positions[i][1] - 1
            boxes[positions[i][0]-1] = 0
    else:
        if positions[i][0] == 1:
            boxes[positions[i][0]] += boxes[positions[i][0]-1] - positions[i][1]
            boxes[positions[i][0]-1] = 0
        elif positions[i][0] == n:
            boxes[positions[i][0]-1-1] += positions[i][1] - 1
            boxes[positions[i][0]-1] = 0
        else:
            boxes[positions[i][0]-1-1] += positions[i][1] - 1
            boxes[positions[i][0]] += boxes[positions[i][0]-1] - positions[i][1]
            boxes[positions[i][0]-1] = 0
        
for i in range(n):
    print(boxes[i])
```

# [Juicer](https://codeforces.com/contest/709/my)

# [Carrot Cakes](https://codeforces.com/contest/799/my)

```python
import math

n, t, k, d = map(int, input().split())

one_oven = t * (math.ceil(n / k))
counter = 0

while n > 0:
    counter += 1
    if counter % t == 0:
        n -= k
    if counter > d:
        if (counter - d) % t == 0:
            n -= k

if counter >= one_oven:
    print("NO")
else:
    print("YES")
```
