```python
import numpy as np

def repeated(N, matrix):
    counter = 0
    for row in range(N):
        if N != (len(np.unique(matrix[row]))):
            counter = counter + 1
    return counter
            
T = int(input())

for i in range(T):
    
    N = int(input())
    trace = 0
    matrix = []
    for row in range(N):
        matrix.append([])
        matrix[row].extend(list(map(int, input().split())))
        trace = trace + matrix[row][row]
    
    repeated_rows = repeated(N, matrix)
    repeated_columns = repeated(N, np.array(matrix).T.tolist())
    
    print("case #{}: {} {} {}".format((i+1), trace, repeated_rows, repeated_columns))
```
