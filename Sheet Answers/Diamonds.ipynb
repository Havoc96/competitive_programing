```python
n = int(input())

# Top Line:
print("+" + "-"*(2*n) + "+")

# Upper Middle Lines:
for i in range(n-1):
    if i%2 == 0:
        print("|" + " "*(n-1-i) + "/" + "="*(2*i) + "\\" + " "*(n-1-i) + "|")
    else:
        print("|" + " "*(n-1-i) + "/" + "-"*(2*i) + "\\" + " "*(n-1-i) + "|")

# Middle Line:
if n%2 == 0:
    print("|<" + "-"*(n+(n-2)) + ">|")
else:
    print("|<" + "="*(n+(n-2)) + ">|")

# Bottom Middle Lines:
for i in range(n-1, 0, -1):
    if i%2 == 0:
        print("|" + " "*(n-i) + "\\" + "-"*(2*(i-1)) + "/" +  " "*(n-i) + "|")
    else:
        print("|" + " "*(n-i) + "\\" + "="*(2*(i-1)) + "/" +  " "*(n-i) + "|")

# Bottom Line:
print("+" + "-"*(2*n) + "+")
```
